```python
from time import time 
import pyarrow.parquet as pq
import pandas as pd
import numpy as np
import itertools
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={"figure.figsize":(5,6)})
```

# ALS result visualizations for small dataset

```python
df = pd.read_csv('/content/small-asl.csv')
df["Precision"] = df["Precision"].round(2)
df["Maps"] = df["Maps"].round(3)
df["NDCGs"] = df["NDCGs"].round(2)
df["RMSE"] = df["RMSE"].round(1)
df["Times"] = df["Times"].astype(int)
```

```python
# ALS visualization plot for 'Precision'
graph = sns.FacetGrid(df , col="Iter", hue="Precision")
graph.map(plt.scatter, "Rank" , "Reg_Param"  , edgecolor="w").add_legend()
graph.fig.subplots_adjust(top=0.7)
graph.fig.suptitle("ALS - Precision" , fontsize=16)
plt.show()
```

```python
# ALS visualization plot for 'Maps'
graph = sns.FacetGrid(df , col="Iter", hue="Maps")
graph.map(plt.scatter, "Rank" , "Reg_Param"  , edgecolor="w").add_legend()
graph.fig.subplots_adjust(top=0.7)
graph.fig.suptitle("ALS - Maps" , fontsize=16)
plt.show()
```

```python
# ALS visualization plot for 'NDCGs'
graph = sns.FacetGrid(df , col="Iter", hue="NDCGs")
graph.map(plt.scatter, "Rank" , "Reg_Param"  , edgecolor="w").add_legend()
graph.fig.subplots_adjust(top=0.7)
graph.fig.suptitle("ALS - NDCGs" , fontsize=16)
plt.show()
```

```python
# ALS visualization plot for 'RMSE'
graph = sns.FacetGrid(df , col="Iter", hue="RMSE")
graph.map(plt.scatter, "Rank" , "Reg_Param"  , edgecolor="w").add_legend()
graph.fig.subplots_adjust(top=0.7)
graph.fig.suptitle("ALS - RMSE" , fontsize=16)
plt.show()
```
